### What is the Binomial Tree Model?

The binomial tree model is a model used to evaluate various options including exotic options with complex payoff functions. Given an underlying asset S, we assume that in the next time period, S will be in 1 of 2 possible states. In this case, let us name them the "Up" state and "Down" state corresponding to S increasing or decreasing in value during the next step. We can continue to build these steps to form a "tree", once we've formed a sufficiently large tree, we can use the tree as the model for our the various possible movements of our underlying. With the model, given any option, we can compute the payoffs, discount them given the risk-free interest rate, and weight them which will allow us to compute the option value.

Below we'll start easy with a 2 period model but will then expand



```python
## 2-Period Model Call Option
import numpy as np
import matplotlib as mp
import matplotlib.pyplot as plt
import math
#S
startPrice = 10.0
#K
strikePrice = 10.0
#U
upFactor=1.2
#D
downFactor=.8
#R
riskFreeRate=.05
#Df
discountFactor = 1/(1+riskFreeRate)

#Note: Call option Payoff = Max(S-K, 0)

#At T=2 We'll Have 4 possible states:
suu = startPrice * upFactor * upFactor
sud = startPrice * upFactor * downFactor
sdu = startPrice * downFactor * upFactor
sdd = startPrice * downFactor * downFactor

print("SUU: "+str(suu))
print("SUD: "+str(sud))
print("SDU: "+str(sdu))
print("SDD: "+str(sdd))

#Next We Need To Compute The Probabilites Of An Up-State or Down-State
#We'll Let P,Q denote probability of an up-state movement and down-state movement resp.
#We know that there are 2 possible states, and a probability space must sum to 1 So P+Q = 1
#We also know that the the weighted discounted sum of SU and SD must be S, yielding (1/R * (P*SU + Q*SD)) = S

A = np.array([[1,1],[startPrice*upFactor*discountFactor, startPrice*downFactor*discountFactor]])
B = np.array([1, startPrice])

X = np.linalg.inv(A).dot(B)
upFactorProb = X[0]
downFactorProb=X[1]
#print(X)

print("\n\nProbability of An Up-Movement: "+str(X[0]))
print("Probability of An Down-Movement: "+str(X[1]))


#Now We Need to Compute the Call Option Payoffs
puu = max(suu-strikePrice,0)
pud = max(sud-strikePrice,0)
pdu = max(sdu-strikePrice,0)
pdd = max(sdd-strikePrice,0)

print("\n\nPUU: "+str(puu))
print("PUD: "+str(pud))
print("PDU: "+str(pdu))
print("PDD: "+str(pdd))

#Now All that is left is to discount them

#Time One Option Value

pu = (puu*upFactorProb+pud*downFactorProb)*discountFactor
pd = (pdu*upFactorProb+pdd*downFactorProb)*discountFactor

print("Time One Option Price In Up State: "+str(pu))
print("Time One Option Price In Down State: "+str(pd))

#Time Zero Option Value
p = (pu*upFactorProb + pd*downFactorProb)*discountFactor

print("\n\nTime Zero Option Price: "+str(p))
```

### Cox-Ross Rubenstein Model

The CRR model is a specical form of the Binomial model where the Up/Down factors are computed based on the following formulas. In order to have no-arbitrage arguments hold, we need u>1>d>0,S(0)>0, and r>0.

In this case: 

$$u=e^{\sigma\sqrt{\Delta t}}$$
where sigma represents the volatility of the underlying.

The downward movement factor is simply d=1/u.
We also can find the associated probabilites. The probability of upward movement (P) is simply 

$$P = (e^{r \Delta t} - d) / (u-d)$$
And Q = 1-P.

Using this, we are basically ready to apply the same logic from the example above.
We'll build a more robust example now below

### Jarrow Rudd Model

The Jarrow Rudd model is similar to CRR in that it defines a special computation for the Up/Down factors. In this case Jarrow Rudd define the up-factor to be 

$$u=e^{(r-\sigma^2/2) \Delta t + \sigma\sqrt{\Delta t}}$$

and the down factor to be very similar

$$d=e^{(r-\sigma^2/2) \Delta t - \sigma\sqrt{\Delta t}}$$

One of the neat tricks for Jarrow Rudd is that it turns out our probabilites of Up/Down factors in the model are both .5, meaning there's an equal likelihood of the underlying's value going up or down.

$$ P = Q = 0.5 $$

### Tian Model

The Tian model is another common model similar to JR/CRR that were discussed above. Here again we have specified ways to compute up/down factors, these will then drive how we determine the associated up/down probabilities. Here we first let $$ M = e^{r\Delta t}$$ and let $$V = e^{\sigma^2 \Delta t}$$

We then define the up-factor to be as follows:

$$u=MV/2[(V+1)+\sqrt{V^2+2V-3}]$$

Similarly, we define out down-factor to be $$d=MV/2[(V+1)-\sqrt{V^2+2V-3}]$$

This allows us to compute the up-probability to be $$P=(M-d)/(u-d)$$ and $$Q=1-P$$

```python
def generic_option_value(S0, K, T, r, sigma, m, optionType, modelType):
    dt = T / m
    df = (1 / (1 + (r * dt)))
    factors = getFactors(modelType, dt, sigma, r)
    u = factors[0]
    d = factors[1]
    upFactorProb = factors[2]
    downFactorProb = factors[3]

    #Creates the tree
    mu = np.arange(m + 1)
    mu = np.resize(mu, (m + 1, m + 1))
    md = np.transpose(mu)
    mu = u ** (mu - md)
    md = d ** md
    S = S0 * mu * md
    #print(S)

    z=0
    V = getPayoff(optionType, S, K)
    for t in range(m-1, -1, -1):
        V[0:m-z, t] = (upFactorProb * V[0:m-z, t+1] + downFactorProb * V[1:m-z+1, t+1]) * df
        z+=1
    #print(V[0,0])

    return(V[0,0])


def getFactors(model, dt, sigma, r):
    if model == "CRR":
        #print("Using CRR")
        up = math.exp(sigma * math.sqrt(dt))
        down = 1 / up
        upProb = (math.exp(r * dt) - down) / (up - down)
        downProb = 1 - upProb
    elif model == "Jarrow":
        #print("Using Jarrow Rudd")
        up = math.exp((r-((sigma**2)/2))*dt + sigma*math.sqrt(dt))
        down = math.exp((r-((sigma**2)/2))*dt - sigma*math.sqrt(dt))
        upProb = downProb = .5
    else:
        #print("Using Tian")
        M = math.exp(r*dt)
        V = math.exp(sigma **2 * dt)
        up = (M*V/2)*((V+1)+math.sqrt(V**2 + 2*V - 3))
        down = (M*V/2)*((V+1)-math.sqrt(V**2 + 2*V - 3))
        upProb = (M-down)/(up-down)
        downProb = 1-upProb
    return [up, down, upProb, downProb]
        

def getPayoff(optionType, S, K):
    if optionType == "Call":
        V = np.maximum(S-K, 0)
    else:
        V = np.maximum(K-S, 0)
    return V
```

### Valuation Of the Same Option Using All 3 Discrete Models

Below value the same option for all three models. Note here they are all around $10. The differences come from the up/down factor calculations. 

```python
S0 = 100
K = 100
T = 1
r = 0.05
sigma = .2
m = 10
optionType = "Call"

generic_option_value(S0,K,T,r,sigma,m, "Call", "Jarrow")
generic_option_value(S0,K,T,r,sigma,m, "Call", "CRR")
generic_option_value(S0,K,T,r,sigma,m, "Call", "Tian")
```

### Black Scholes Valuation
The Black Scholes Model is the most common model used for option valuation. We will not go thru the computation here, there is a separate Notebook I'll be sharing for that. However, we'll talk a bit about it. Black Scholes is a continuous time option pricing model. It's the dominant model utilized when valuing path-independent options. While it has some flaws in it's assumptions it's still widely used today across the industry. Given Black Scholes is still widely considered on the Street to be the "Correct" option price, we'll use that as the "official" price of an option.

### Convergence of Binomial Models to BSM
We'll now show graphically how as we continue to shrink our dt by increasing the number of intervals, the price will begin to converge to BSM regardless of what model we use.

```python
#Call Option Example
crrPrices=[]
jrPrices=[]
tianPrices = []
index = []

S0 = 100
K = 100
T = 1
r = 0.05
sigma = .2
m = 1
optionType = "Call"

black_scholes_price = 10.45

for i in range(0,100):
    index.append(i)
    crrPrices.append(generic_option_value(S0,K,T,r,sigma,i+1, optionType, "CRR"))
    jrPrices.append(generic_option_value(S0,K,T,r,sigma,i+1, optionType, "Jarrow"))
    tianPrices.append(generic_option_value(S0,K,T,r,sigma,i+1, optionType, "Tian"))

plt.figure(figsize=(12,8))
plt.plot(index, crrPrices, label = "CRR")
plt.plot(index, jrPrices, label = "JR")
plt.plot(index, tianPrices, label = "Tian")
plt.axhline(black_scholes_price, color="r", ls="dashed", lw=1.5, label="Black-Scholes")
plt.grid()
plt.xlabel("# Of Steps")
plt.ylabel("European Call Option Value")
plt.legend(loc=4)


```

```python
#Put Option Example

crrPrices=[]
jrPrices=[]
tianPrices = []
index = []

S0 = 100
K = 100
T = 1
r = 0.05
sigma = .2
m = 1
optionType = "Put"

black_scholes_price = 5.57

for i in range(0,100):
    index.append(i)
    crrPrices.append(generic_option_value(S0,K,T,r,sigma,i+1, optionType, "CRR"))
    jrPrices.append(generic_option_value(S0,K,T,r,sigma,i+1, optionType, "Jarrow"))
    tianPrices.append(generic_option_value(S0,K,T,r,sigma,i+1, optionType, "Tian"))

plt.figure(figsize=(12,8))
plt.plot(index, crrPrices, label = "CRR")
plt.plot(index, jrPrices, label = "JR")
plt.plot(index, tianPrices, label = "Tian")
plt.axhline(black_scholes_price, color="r", ls="dashed", lw=1.5, label="Black-Scholes")
plt.grid()
plt.xlabel("# Of Steps")
plt.ylabel("European Put Option Value")
plt.legend(loc=4)
```

### Conclusion

From what we see above, as we begin to shrink the interval size we see that regardless of the option type and what model we use, all three begin to converge to Black Scholes model option price. Binomial models are still extremely useful in the analysis and valuation of both path-dependent and path-independent models.
